# Biological Growth -- Coyotes

```python
import numpy as np 
import matplotlib.pyplot as plt
```

# Exponential Growth

```python
nsteps = 100 #Number of steps in our stimulation
x = np.zeros(nsteps) #Number of animals at every step 
x[0] = 1 #The original number of animals 
r = 1.1 #Rate of population change (birth rate - death rate), if above 1 the pop will grow, if below 1 they die out

for i in range (1,nsteps):
    x[i] = x[i-1] * r  #Calculation

plt.plot(x); #Plot of the calculation
plt.xlabel('Number of generations');
plt.ylabel('Population');
```

# Logistic Growth 

```python
nsteps = 100 #Number of steps in our stimulation
x = np.zeros(nsteps) #Number of animals at every step 
x[0] = 0.6 #The original population size
r = 3.4 #Rate of population change (birth rate - death rate), if above 1 the pop will grow, if below 1 they die out
      #Also the speed of reaction
      #System overreacts at r=3, tries to get to the stable point
      #Can come to a stable point, can oscillate, can die out
      #When r=4, wild behavior, almost dies out at 80 (x-axis)
      #When r=3, oscillation; when r=3.5 mid-oscillation; r=4, gets crazy

for i in range (1,nsteps):
    x[i] = r*x[i-1] * (1-x[i-1])  #Calculation

plt.plot(x); #Plot of the calculation
plt.ylim(0,1) #Force y axis to be between 0 and 1
```
