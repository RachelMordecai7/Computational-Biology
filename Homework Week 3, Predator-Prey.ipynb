```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 50000
a = 1.01 #Bunny reproduction rate
b = 0.89 #Bunny predation rate 
c = 0.6  #Conversion rate 
d = 1    #Fox death rate
e = 0.67 #Fox predation by wolves 

dt = 0.001 #For graphical purposes

x = np.zeros(nsteps)
y = np.zeros(nsteps)

fig1 = plt.figure(figsize=(16,5));
fig2 = plt.figure(figsize=(5,5));
ax2 = plt.axes()

possible_es = [0.6, 1, 1.3, 1.9] #Here (as explained below) I would have also liked to vary the bunny predation rate based on the number of foxes.

for itrial in range(len(possible_es)): 
    e = possible_es[itrial]
    x[0] = 1
    y[0] = 1
    
    for t in range(1,nsteps):
        x[t] = x[t-1] + dt * (a * x[t-1] - b * x[t-1] * y[t-1])
        y[t] = y[t-1] + dt * (c * x[t-1] * y[t-1] - d * y[t-1]) * e #Changed this equation by multiplying by the predation rate of foxes.
        
        
    ax1 = fig1.add_subplot(1, len(possible_es), itrial + 1);
    ax1.plot(x, label = "bunnies");
    ax1.plot(y, label = "foxes");
    ax1.legend(loc = "upper right");
    
    ax2.plot(x, y, label = e);
ax2.legend(loc = "upper right");       

```

* Here I attempted to add a predation rate to the population of foxes. I named the variable e, and created a range of possible values of e, then graphed them. When the predation rate of foxes increases, we can see a greater variability and larger fluctuation of the populations of both bunnies and foxes when the predators of foxes (wolves) are present (assuming I did it correctly). This might have the effect of making the populations less stable.
* In order to add in this factor, I multiplied the equation for the population of foxes by the predation rate of the foxes. 
* This equation is also assuming that regardless of the amount of foxes, the predation rate of bunnies stays the same, I was looking for a way to vary this and to have the predation rate of bunnies vary based on the number of foxes, so I tried to add in another range of possible values but for b.
