# Simple numerical integration for motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Attempt 1: Throwing vertical

```python
a = -1 #Acceleration, negative because it tries to reverse the flight of the stone, this is the parameter of the model
v = 5 #We throw the stone upwards, so it is positive
x = 0 #We throw it from the surface so x=0
dt = 0.01 #Our step, numerical integration step is supposed to be small
nsteps = 1000
history = np.zeros(nsteps) #We will plot this eventually

for i in range (nsteps):
    v = v + a*dt
    x = x + v*dt
    history[i] = x

plt.plot(history);
    
```

# Attempt 2: Throwing at an angle

```python
a = -1 #Acceleration, negative because it tries to reverse the flight of the stone
vy = 5 # Vertical component of the speed, and it's upwards => positive
vx = -1 # Horizontal component of the speed 
x = 0 #We throw it from the house, and are trying to see how far it goes
y = 0 #We throw it from the surface so y=0
dt = 0.01 #Our step, numerical integration step is supposed to be small, this is the parameter of our appraoch to modeling
nsteps = round(10/dt)

for i in range (nsteps):
    vy = vy + a*dt #Update vertical speed
    
    x = x + vx*dt #Update x position 
    y = max(y + vy*dt , 0) #Update y position
    
    plt.plot(x,y,'k.') #Just a new plot, instead of remembering x and y, (lazy)
```

# Attempt 3: Throwing at an angle with air friction

```python
a = -1 
vy = 5 
vx = 1 
x = 0 
y = 0 
f = 0.01 #How important friction is relative to gravity

dt = 0.01 
nsteps = round(10/dt)

for i in range (nsteps):
    vy = vy + a*dt - vy*f #Updated vy taking into account air friction
    vx = vx - vx*f #Updated vx taking into account air friction
    
    x = x + vx*dt 
    y = max(y + vy*dt , 0) 
    
    plt.plot(x,y,'k.') 
    
    
```
